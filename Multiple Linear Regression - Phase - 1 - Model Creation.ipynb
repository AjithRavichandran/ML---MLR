```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv('50_Startups.csv')
```

```python
dataset
```

```python
dataset=pd.get_dummies(dataset,dtype=int, drop_first=True) # this code get_dummies is used to convert NOMINAL characterstic into numeric and drop_first will helps to delete the extra column like --> California
```

```python
dataset.columns
```

```python
independent=dataset[['R&D Spend','Administration','Marketing Spend','State_Florida','State_New York']]
```

```python
independent
```

```python
dependent=dataset[['Profit']]
dependent
```

```python
plt.figure(figsize=(4,4))
plt.plot(independent,dependent)
plt.xlabel('Investment',fontsize=15)
plt.ylabel('Profit',fontsize=15)
plt.show()
```

```python
from sklearn.model_selection import train_test_split
input_train,input_test,output_train,output_test=train_test_split(independent,dependent,test_size=1/3,random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(input_train,output_train)
```

```python
weight=regressor.coef_
bias=regressor.intercept_
print(f'Weight is {weight}')
print(f'Bias is {bias}')
```

```python
predict=regressor.predict(input_test)
```

```python
from sklearn.metrics import r2_score
r2=r2_score(output_test,predict)
r2
```

```python
import pickle
filename='MLR Model.sav'
pickle.dump(regressor,open(filename,'wb'))
```

```python
load=pickle.load(open('MLR Model.sav','rb'))
```
